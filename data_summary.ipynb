```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from datetime import timedelta, datetime

%matplotlib inline
```

```python
pco = pd.read_csv("profile_cutoff.csv")
```

```python
colAnaList = ['diff_temp','vgrad']

for col in colAnaList:
    for limit in [10,20]:
        pco["min_{0}_{1}".format(limit,col)] = pco[col].rolling(window=limit,min_periods=0).min().round(decimals=3)
        pco["max_{0}_{1}".format(limit,col)] = pco[col].rolling(window=limit,min_periods=0).max().round(decimals=3)
        pco["var_{0}_{1}".format(limit,col)] = pco[col].rolling(window=limit,min_periods=0).var().round(decimals=3)
        pco["std_{0}_{1}".format(limit,col)] = pco[col].rolling(window=limit,min_periods=0).std().round(decimals=3)
        pco["avg_{0}_{1}".format(limit,col)] = pco[col].rolling(window=limit,min_periods=0).mean().round(decimals=3)
        pco["norm_{0}_{1}".format(limit,col)] = round(
            (abs(pco[col] - pco["avg_{0}_{1}".format(limit,col)]) / pco["std_{0}_{1}".format(limit,col)]),3)
        
    for nexnum in np.arange(-1,-6,-1):
        pco["next_{0}_{1}".format(abs(nexnum),col)] = pco[col].shift(nexnum)

    pco["std_next_5_{0}".format(col)] = pco.loc[:,["next_1_{0}".format(col),"next_2_{0}".format(col),"next_3_{0}".format(col),"next_4_{0}".format(col),"next_5_{0}".format(col)]].std(axis=1).round(decimals=3)
    pco["avg_next_5_{0}".format(col)] = pco.loc[:,["next_1_{0}".format(col),"next_2_{0}".format(col),"next_3_{0}".format(col),"next_4_{0}".format(col),"next_5_{0}".format(col)]].mean(axis=1).round(decimals=3)
    pco["std_next_4_{0}".format(col)] = pco.loc[:,["next_1_{0}".format(col),"next_2_{0}".format(col),"next_3_{0}".format(col),"next_4_{0}".format(col)]].std(axis=1).round(decimals=3)
    pco["avg_next_4_{0}".format(col)] = pco.loc[:,["next_1_{0}".format(col),"next_2_{0}".format(col),"next_3_{0}".format(col),"next_4_{0}".format(col)]].mean(axis=1).round(decimals=3)
    pco["std_next_3_{0}".format(col)] = pco.loc[:,["next_1_{0}".format(col),"next_2_{0}".format(col),"next_3_{0}".format(col)]].std(axis=1).round(decimals=3)
    pco["avg_next_3_{0}".format(col)] = pco.loc[:,["next_1_{0}".format(col),"next_2_{0}".format(col),"next_3_{0}".format(col)]].mean(axis=1).round(decimals=3)
    pco["std_next_2_{0}".format(col)] = pco.loc[:,["next_1_{0}".format(col),"next_2_{0}".format(col)]].std(axis=1).round(decimals=3)
    pco["avg_next_2_{0}".format(col)] = pco.loc[:,["next_1_{0}".format(col),"next_2_{0}".format(col)]].mean(axis=1).round(decimals=3)
```

```python
pco['last_point'] = (pco['number'] == 1449).astype(int)
```

```python
pco['away_from_last_point'] = pco['number'] - 1449
```
